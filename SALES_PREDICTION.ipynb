Importing libraries and loading the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score

advertising = pd.DataFrame(pd.read_csv("advertising.csv"))
```

Showing the correlation between all features and the target, as we see the TV has the most correlation, so we will make a linear regression model using only the TV

```python
sns.heatmap(advertising.corr(), cmap="YlGnBu", annot = True)
plt.show()
```

Preparing the feature and the target

```python
X = advertising[['TV']]
y = advertising['Sales']
```

splitting our data into training and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

initializing our model and fitting our training data 

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

testing our model on the test data

```python
y_pred = model.predict(X_test)
```

Calculating the mean squared error

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

plotting our predicted line on the test data

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred,color = 'red')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

plotting our predicted line on the train data

```python
plt.scatter(X_train, y_train)
plt.plot(X_test, y_pred,color = 'red')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```
